```python
#2 a. Prompt the user to enter the name of a product that a customer wants to order.
#b.Add the product name to the "customer_orders" set.
#c.Ask the user if they want to add another product (yes/no)
#d. continue the loop until the user does not want to add anoother product
customer_orders = set ()
while True :
    product = input ("Which product would you like to checkout? (separated by comma): T-shirt, Mug, Book, Keychain, Hat").lower()

    if product in inventory:
        customer_orders.add(product)
        print (f"{product}was added to your order")
        
    else:
       print (f"Sorry, {product} is not available")

    customer_input = input(f"Would you like to checkout another product? yes/no").lower()
        
    if customer_input == "no":
        print (f"Thank you. We've received your order, {customer_orders}!")
        break 
```

```python
#3Instead of updating the inventory by subtracting 1 from the quantity of each product, 
#only do it for the products that were ordered (those in "customer_orders").
for product_key in customer_orders:
    if product_key in inventory and inventory[product_key] > 0:
        inventory[product_key] -= 1
print(inventory)
```
